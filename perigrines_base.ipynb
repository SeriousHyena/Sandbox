# Import Required Libraries and Modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


%matplotlib inline
```

### Read in the datafile containing coordinates of the Perigrine sightings and assign the data to a dataframe for use in our program

```python
df = pd.read_csv(r'C:\Users\emack\Documents\Sandbox\pfunique2.csv')

# in order to update the perigrine presences on the basemap simply read-in the new .csv file between the ' ' above. be sure to use explicit pathing per 
# the file location on your system and basemap and location ticks will update once the notebook cells are executed again.
```

### Loop through the dataframe and assign the coordinates to their respective variables for use in the map implementation

```python
lons = [] #variable for the longitudes
lats = [] #variable for latitudes
counter = 0 #a simple counter to indicate the number of lat/lon entries

for index, row in df.iterrows(): #iterate through the dataframe and get lat/lon
# access data using column names
    long = row['longitude']
    latt = row['latitude']
    lons.append(long) #append the coordinate to the variable
    lats.append(latt) #append the coordinate to the variable
    counter += 1 #increment the counter
    
```

### Plot the base map and add the Perigrine sightings to the basemap

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 40})

fig, ax = plt.subplots(figsize=(50,25))
map = Basemap(width=12000000,height=9000000)
# the continents will be drawn on top.


ax.set(title = "Global Distribution of Perigrine Falcons")
ax.text(-50, -50, 'Magenta Ticks Indicate Perigrine Presence', fontsize=35)

map.drawcoastlines()
# the continents will be drawn on top.
x, y = map(lats, lons)
map.scatter(x, y, marker='D',color='m')


# draw parallels and meridians.
# label parallels on right and top
# meridians on bottom and left
parallels = np.arange(-81.,81,10.)
# labels = [left,right,top,bottom]
map.drawparallels(parallels,labels=[False,True,True,False])
meridians = np.arange(10.,351.,20.)
map.drawmeridians(meridians,labels=[True,False,False,True])
map.drawstates(linewidth=0.5, linestyle='solid', color='k', antialiased=1, ax=None, zorder=None) #show state borders
map.drawcountries(linewidth=0.5, linestyle='solid', color='k', antialiased=1, ax=None, zorder=None) #show country borders

#deprecation warnings can be safely ignored.

plt.show()
```

# Basemap with Perigrine sightings plotted.
